For example, if a user enters 1001300110 and specified the base to be 2, this should give an error because the digit '3' cannot be in a binary number. So, we will write a function to check that a number is valid in a given base system

```python
def check_number(n, b):
    '''
    Check that a number n is indeed in base b 
    by making sure that this number has no digits out of the range of the base
    
    Parameters:
    n: number to be checked
    b: base to be checked against
    
    Returns:
    True if the number is valid in this base, False otherwise
    '''
    valid_digits = []
    for i in range(b):
        valid_digits.append(digit(i))
    for d in list(str(n)):
        if d not in valid_digits:
            return False
    return True
```
